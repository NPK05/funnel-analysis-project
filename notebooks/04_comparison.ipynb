```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np

SAVE_PATH = '../outputs/dashboard/'

print("✓ Comparison notebook ready")
```

```python
# CROSS-DATASET FUNNEL SHAPE COMPARISON
# Shows how each funnel archetype looks differently

fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=[
        'Google Merch<br><sup>Conversion Funnel</sup>',
        'Instacart<br><sup>Retention Funnel</sup>',
        'Olist<br><sup>Fulfilment Funnel</sup>'
    ],
    horizontal_spacing=0.08
)

# ── Google Merch ──
gm_stages = ['Add to Cart','Checkout','Purchase']
gm_counts = [12545, 6404, 4066]
gm_pcts   = [100.0, 51.0, 32.4]
gm_colors = ['#378ADD','#EF9F27','#1D9E75']

for i, (stage, count, pct, color) in enumerate(
    zip(gm_stages, gm_counts, gm_pcts, gm_colors)
):
    fig.add_trace(go.Bar(
        x=[stage], y=[pct],
        marker_color=color, opacity=0.85,
        text=f'<b>{pct:.0f}%</b><br>{count:,}',
        textposition='inside',
        insidetextanchor='middle',
        textfont=dict(size=11, color='white'),
        showlegend=False
    ), row=1, col=1)

# ── Instacart ──
ic_stages = ['Active (3-5)','Loyal (6+)']
ic_counts = [206209, 156805]
ic_pcts   = [100.0, 76.0]
ic_colors = ['#378ADD','#1D9E75']

for stage, count, pct, color in zip(
    ic_stages, ic_counts, ic_pcts, ic_colors
):
    fig.add_trace(go.Bar(
        x=[stage], y=[pct],
        marker_color=color, opacity=0.85,
        text=f'<b>{pct:.0f}%</b><br>{count:,}',
        textposition='inside',
        insidetextanchor='middle',
        textfont=dict(size=11, color='white'),
        showlegend=False
    ), row=1, col=2)

# ── Olist ──
ol_stages = ['Created','Approved','Shipped','Delivered','Reviewed']
ol_counts = [99441, 99281, 97658, 96476, 95832]
ol_pcts   = [100.0, 99.8, 98.2, 97.0, 96.4]
ol_colors = ['#378ADD','#1D9E75','#1D9E75','#1D9E75','#534AB7']

for stage, count, pct, color in zip(
    ol_stages, ol_counts, ol_pcts, ol_colors
):
    fig.add_trace(go.Bar(
        x=[stage], y=[pct],
        marker_color=color, opacity=0.85,
        text=f'<b>{pct:.1f}%</b><br>{count:,}',
        textposition='inside',
        insidetextanchor='middle',
        textfont=dict(size=10, color='white'),
        showlegend=False
    ), row=1, col=3)

# key metric annotations per dataset
fig.add_annotation(
    x=1, y=-18,
    text='<b>Key drop: ATC→Checkout 49%</b><br>$552K revenue leakage',
    showarrow=False,
    font=dict(size=11, color='#D85A30'),
    bgcolor='rgba(255,255,255,0.9)',
    bordercolor='#D85A30', borderwidth=1,
    xref='x', yref='y'
)

fig.add_annotation(
    x=0.5, y=-18,
    text='<b>Key drop: Active→Loyal 24%</b><br>Weekly users worth 5x monthly',
    showarrow=False,
    font=dict(size=11, color='#D85A30'),
    bgcolor='rgba(255,255,255,0.9)',
    bordercolor='#D85A30', borderwidth=1,
    xref='x2', yref='y2'
)

fig.add_annotation(
    x=2, y=-18,
    text='<b>Key issue: 8.1% late rate</b><br>R$1.85M revenue at risk',
    showarrow=False,
    font=dict(size=11, color='#D85A30'),
    bgcolor='rgba(255,255,255,0.9)',
    bordercolor='#D85A30', borderwidth=1,
    xref='x3', yref='y3'
)

fig.update_layout(
    title=dict(
        text='Three Funnel Archetypes — Side by Side Comparison<br>'
             '<sup>Conversion (Google) vs Retention (Instacart) vs Fulfilment (Olist)</sup>',
        x=0.5, font=dict(size=18, color='#2E4057')
    ),
    barmode      = 'group',
    font         = dict(family='Arial', size=12),
    plot_bgcolor = 'white',
    paper_bgcolor= 'white',
    height=580, width=1200,
    showlegend=False,
    margin=dict(b=120)
)

fig.update_yaxes(
    range=[0,115], ticksuffix='%',
    gridcolor='#F0F0F0', title='% of funnel entry'
)

fig.write_html(SAVE_PATH + '15_cross_dataset_comparison.html')
fig.show()
print("✓ Saved: 15_cross_dataset_comparison.html")
```
