## The Five Cross-Dataset Insights

### 1 Funnel archetype determines the right metric
Google Merch CVR 32.4% is not bad it is expected for a conversion
funnel where most users browse without purchase intent. Olist 97%
fulfilment is not perfect the real story is delivery quality among
the 97%. Instacart 76% loyalty is exceptionally strong for a delivery
platform. Each funnel must be benchmarked against its own archetype,
not against each other.

### 2 Product type drives behavior pattern
- Google Merch peaks Wednesday (office browsing, branded merchandise)
- Instacart peaks Saturday 10am (weekly grocery planning routine)
- Same analytical approach, completely opposite conclusions
- Never assume findings from one product type apply to another

### 3 Drop-off cause determines the fix
| Dataset | Drop-off type | Root cause | Right intervention |
|---|---|---|---|
| Google Merch | User decision | UX/price friction | Reduce checkout friction |
| Instacart | Habit failure | Cadence not formed | Nudge at order 4-5 |
| Olist | Operational failure | Seller/logistics | Accountability + investment |

### 4 Sample size determines what questions you can answer
- Google Merch country differences: p>0.05 NOT significant
  (only ~200 users per country insufficient to detect differences)
- Instacart cadence differences: p=0.000000
  (200K users enormous statistical power)
- Olist geographic differences: p=0.000000
  (40K SP orders enough to detect even small differences)

Sample size is not just a data quality issue.
It determines which hypotheses are testable.

### 5 The single biggest opportunity per dataset
| Dataset | Opportunity | Metric | Impact |
|---|---|---|---|
| Google Merch | Fix cart→checkout CTA | 49% drop-off | $400K leakage |
| Instacart | Convert monthly→bi-weekly | 22pp loyalty gap | 5x LTV difference |
| Olist | Fix RJ delivery reliability | 12K orders, 13.5% late | ~R$270K direct |
